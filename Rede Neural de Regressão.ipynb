```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
dados = pd.DataFrame({
    'celsius': [-40, -10, 0, 8, 15],
    'fahrenheit': [-40, 14, 32, 46, 59]
})
```

```python
dados
```

```python
X = dados['celsius'].values
y = dados['fahrenheit'].values
```

```python
modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])

modelo.compile(
    optimizer=tf.keras.optimizers.Adam(.1),
    loss='mean_squared_error'
)
```

```python
treinamento = modelo.fit(X, y, epochs=1000)
```

```python
plt.xlabel("# Epoch")
plt.ylabel("Magnitude da perda")
plt.plot(treinamento.history['loss'])
```

```python
c = np.array([[1230]])
predicao = modelo.predict(c)
predicao
```
