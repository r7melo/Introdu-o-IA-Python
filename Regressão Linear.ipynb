## Separar caracteristicas de para treinamento

```python
X = dados_na.drop(['median_house_value'], axis=1)
y = dados_na['median_house_value']
```

```python
from sklearn.model_selection import train_test_split

X_ent, X_pru, y_ent, y_pru = train_test_split(X, y, test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_ent, y_ent)
```

```python
predicao = model.predict(X_pru)
```

```python
comparacao = pd.DataFrame({
    'Predição': predicao,
    'Real': y_pru
})

comparacao
```

```python
# Overfitting sobreajuste
print(model.score(X_ent, y_ent))
print(model.score(X_pru, y_pru))
```

```python
# Medição de erros

from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_pru, predicao)
```

```python
mse
```

```python
rmse = np.sqrt(mse)
```
